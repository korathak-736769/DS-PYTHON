```python
# create class
class Employee:
    def detail(self,emp_name,emp_salary,emp_department): #create method
        self.name = emp_name
        self.salary = emp_salary
        self.department = emp_department
    def Show_Data(self):
        print("Name_Employee : {}".format(self.name))
        print("Salary_Employee : {} Bath".format(self.salary))
        print("Department_Employee : {}".format(self.department))
        
# create objectc 
obj1 = Employee()
obj1.detail("Korathank",56880,"Programmer")
obj2 = Employee()
obj2.detail("Sompong",34534,"Editor")
obj3 = Employee()
obj3.detail("Saksuk",32443,"Manager")

obj1.Show_Data()
print()
obj2.Show_Data()
print()
obj3.Show_Data()
```

### Constructor & Destructor

```python
""" ["Constructor"] This is a special method that will be used when initializing an object.
create class 

    ["Destrucctor"] is a special method that is opposite of constructor. 
'End of class execution' or is done before object decomposition.
"""
class Employee:
    def __init__(self):
        print("default Constructor")
        
    def __init__(self,emp_name,emp_salary,emp_department):
        self.name = emp_name
        self.salary = emp_salary
        self.department = emp_department
    
    def Show_Data(self):
        print("Name_Employee : {}".format(self.name))
        print("Salary_Employee : {} Bath".format(self.salary))
        print("Department_Employee : {}".format(self.department))
    
    def __del__(self):
        print("Deconstructor")
# create objectc 
obj1 = Employee("Korathank",56880,"Programmer")
obj1.salary = 27000
obj1.name = "Korathank Kongpanna"
obj2 = Employee("Sompong",34534,"Editor")
obj3 = Employee("Saksuk",32443,"Manager")

obj1.Show_Data()
print()
obj2.Show_Data()
print()
obj3.Show_Data()
```

### Encapsulation

```python
# create class
class Employee:
#     def __init__(self,emp_name,emp_salary,emp_department):
#         #public attribute
#         self.name = emp_name
#         self.salary = emp_salary
#         self.department = emp_department

#     def __init__(self,emp_name,emp_salary,emp_department):
#         #protected attribute
#         self._name = emp_name
#         self._salary = emp_salary
#         self._department = emp_department

    def __init__(self,emp_name,emp_salary,emp_department):
        #private attribute
        self._name = emp_name
        self.__salary = emp_salary
        self.__department = emp_department
        
    #protected method
    def _showData(self):
        print("Name_Employee : {}".format(self._name))
        print("Salary_Employee : {} Bath".format(self.__salary))
        print("Department_Employee : {}".format(self.__department))
        
# protected test
# It can be modified but _ must be defined to be accessed within the class.
print("################################################")
obj1 = Employee("Korathank",56880,"Programmer")
obj1.name = "Somsak"
obj1._showData()
print()
obj1._name = "Somsak"
obj1._showData()
print("################################################")

# private Test
# Data can only be edited within a class.
obj1 = Employee("Korathank",56880,"Programmer")
obj1.__salary = 4353
obj1._showData()
print("################################################")
```

#### Getter , Setter Method

```python
class Employee:
    #private attribute
    def __init__(self,emp_name,emp_salary,emp_department):
        self._name = emp_name
        self.__salary = emp_salary
        self.__department = emp_department
        
    #protected method
    def _showData(self):
        print("Employee_Name : {}".format(self._name))
        print("Salary : {}".format(self.__salary))
        print("Department : {}".format(self.__department))
    # setter method 
    def set_Name(self,new_name):
        self._name = new_name
    def set_salary(self,new_salary):
        self.__salary = new_salary
    def set_department(self,new_department):
        self.__department = new_department
        
    # getter method
    def get_name(self):
        return self._name
    def get_salary(self):
        return self.__salary
    def get_department(self):
        return self.__department
    

obj1 = Employee("Korathank",56880,"Programmer")
print(obj1._showData()) 
```

### Inheritance

```python
class Employee:
    #class variable
    _max_salary = 50000
    __min_salary = 15000
    _company_name = 'umbrella'
    def __init__(self,emp_name,emp_salary,emp_department):
        self.__name = emp_name
        self.__salary = emp_salary
        self._department = emp_department
        
    #protected method
    def _showData(self):
        print("Employee_Name : {}".format(self.__name))
        print("Salary : {}".format(self.__salary))
        print("Department : {}".format(self._department))
class Accounting(Employee):
    def __init__(self):
        pass

class Programmer(Employee):
    def __init__(self):
        pass
class Sale(Employee):
    def __init__(self):
        pass

acount = Accounting()
programmer = Programmer()
sale = Sale()

print(acount._max_salary)
print(programmer._company_name)
print(sale._company_name)
```

#### keyword super

```python
# when you want to use like [Contructor,Method,Attribute] in mother class
# To use the mother class, specify keywords [ super ].
class Employee:
    #class variable
    _max_salary = 50000
    __min_salary = 15000
    _company_name = 'umbrella'
    def __init__(self,emp_name,emp_salary,emp_department):
        self.__name = emp_name
        self.__salary = emp_salary
        self._department = emp_department
        
    #protected method
    def _showData(self):
        print("Employee_Name : {}".format(self.__name))
        print("Salary : {}".format(self.__salary))
        print("Department : {}".format(self._department))
class Accounting(Employee):
    _department_name = 'accounting department'
    def __init__(self,name,salary):
        super().__init__(name,salary,self._department_name)
        super()._showData()

class Programmer(Employee):
    _department_name = 'System Development Department'
    def __init__(self,name,salary):
        super().__init__(name,salary,self._department_name)

class Sale(Employee):
    _department_name = 'Sales Department'
    def __init__(self,name,salary):
        super().__init__(name,salary,self._department_name)


acount = Accounting("Sompong",43500)
programmer = Programmer("Jhon",18900)
sale = Sale("Diva",67900)

print()
programmer._showData()
print()
sale._showData()
```

#### Overriding method

```python
# The same text but different internal processes is known as polymorhism.
# keyword [Overloading method & Overriding method]
class Employee:
    #class variable
    _max_salary = 50000
    __min_salary = 15000
    _company_name = 'umbrella'
    def __init__(self,emp_name,emp_salary,emp_department):
        self.__name = emp_name
        self.__salary = emp_salary
        self._department = emp_department
        
    # annual income
    def _get_Income(self):
        return self.__salary*12
    
    def _get_Income(self,bonus=0,overtime=0):
        return (self.__salary*12)+bonus+overtime
    
    def __str__(self):
        return ("Employee_name : {} \nSalary : {} \nDepartment : {} \nAnnual_Income : {}".format(self.__name,self.__salary,self._department,self._get_Income()))
    
    #protected method
    def _showData(self):
        print("Employee_Name : {}".format(self.__name))
        print("Salary : {}".format(self.__salary))
        print("Department : {}".format(self._department))
        
        
class Accounting(Employee):
    _department_name = 'accounting department'
    def __init__(self,name,salary,age):
        super().__init__(name,salary,self._department_name)
        self._age = age
        
    def _showData(self):
        super()._showData()
        print("Age : {}".format(self._age))        
    
    # Another form can be written as follows: 
    def __str__(self):
        return super().__str__()+"\nage : {}\n".format(self._age)
    
class Programmer(Employee):
    _department_name = 'System Development Department'
    def __init__(self,name,salary,experience,skill):
        super().__init__(name,salary,self._department_name)
        self.__exp = experience
        self.__skill = skill
        
    def _showData(self):
        super()._showData()
        print("Experience : {}".format(self.__exp))  
        print("Skill : {}".format(self.__skill))  
        
    def __str__(self):
        return super().__str__()+ "\nExperience : {} \nSkill : {}\n".format(self.__exp,self.__skill)
        
class Sale(Employee):
    _department_name = 'Sales Department'
    def __init__(self,name,salary,area):
        super().__init__(name,salary,self._department_name)
        self._area = area

    def _showData(self):
        super()._showData()
        print("Area : {}".format(self._area))
        
    def __str__(self):
        return super().__str__()+"\nArea : {}\n".format((self._area))
        
acount = Accounting("Sompong",43500,24)
programmer = Programmer("Jhon",18900,'5 year','Python , Php , Java')
sale = Sale("Diva",67900,'Thailand')

# print(acount.__str__())
# print(programmer.__str__())
# print(sale.__str__())

print("Department_Acount     [Annual_Income] : {}".format(acount._get_Income(999)))
print("Department_Programmer [Annual_Income] : {}".format(programmer._get_Income(2342,500)))
print("Department_Sale       [Annual_Income] : {}".format(sale._get_Income()))
```
