##### Create Series from List & Tuple

```python
import pandas as pd 
data_ls = [10,20,'korathank',18.7,True] #list
data_tp = (10,20,'korathank',18.7,False) #tuple
list_sr = pd.Series(data_ls)
tuple_sr = pd.Series(data_tp)
print(list_sr,'\n')
print(tuple_sr)
```

##### Create Series from Numpy

```python
import pandas as  pd
import numpy as np
data_np = np.array([1,2,3,4,5])
numpy_sr = pd.Series(data_np)
numpy_sr
```

##### Create Series Define Index

```python
import pandas as pd
item = ['banana','mango','orange']
index = [34,78,44]
ps = pd.Series(item,index=index)
ps
```

##### Create Series from Dictionary

```python
import pandas as pd
data = {'grape':50,'banana':20,'papaya':30} #dictionary
ps = pd.Series(data)
ps
```

##### Accessing data in a series

```python
import pandas as pd
data = {'grape':50,'banana':20,'papaya':30} #dictionary
ps = pd.Series(data)
print(ps,'\n')
print(ps['grape'])
```

##### Access data in Series (Slice)

```python
import pandas as pd
data = {'grape':50,'banana':20,'papaya':30} #dictionary
ps = pd.Series(data)
print(ps,'\n')
print(ps[1:3])
```

##### Create DataFrame from List and Tuple

```python
import pandas as pd 
data_ls = [10,20,'korathank',18.7,True] #list
data_tp = (10,20,'korathank',18.7,False) #tuple
list_df = pd.DataFrame(data_ls)
tuple_df = pd.DataFrame(data_tp)
print(list_df,'\n')
print(tuple_df,'\n')

# Create a Column-Defined DataFrame
data_age = [13,45,76,34]
age_cols = ['age']
df = pd.DataFrame(data_age,columns=age_cols)
df
```

##### Create a DataFrame from a multidimensional List

```python
import pandas as pd
data = [['Keyboard','computer equipment',3890],
        ['Doll','toy',980],
        ['Iphone 12','Mobile',38900]]
cols = ['Name','Category','Price']
df =  pd.DataFrame(data,columns=cols)
df
```

##### Create DataFrame from List with zip

```python
import pandas as pd
item1 = ['Keyboard','Doll','Iphone 12']
item2 = ['computer equipment','toy','Mobile']
item3 = [3890,980,34690]
data = list(zip(item1,item2,item3))
cols = ['Name','Category','Price']
df = pd.DataFrame(data,columns=cols)
df
```

##### Create a DataFrame from a Dictionary

```python
import pandas as pd
data = [{'Name':'Keyboard','Category':'computer equipment','Price':3890},
        {'Name':'Doll','Category':'toy','Price':980},
        {'Name':'Iphone 12','Category':'Mobile','Price':34588}]
df = pd.DataFrame(data)
df.set_index(['Name'],inplace=True)
df
```

##### Create a DataFrame from Series

```python
import pandas as pd
names = ['Korathank','Sompong','Somsak']
ages = [19,24,45]
scores = [87,34,67]
cols = ['Name','Age','Score']

name_series = pd.Series(names)
age_series = pd.Series(ages)
score_series = pd.Series(scores)

frame = {'Name':name_series,
        'Age':age_series,
        'Score':score_series}

df = pd.DataFrame(frame)
df
```
